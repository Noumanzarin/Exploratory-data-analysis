```python
import pandas as pd

data = {
    'Age':       [23, 31, 19, 45, 52],
    'Salary':    [48000, 54000, 31000, 69000, 72000],
    'Gender':    ['M', 'F', 'F', 'M', 'F'],
    'Purchased': ['Yes', 'No', 'Yes', 'No', 'Yes']
}
df = pd.DataFrame(data)
```

# Univariate

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Histogram of Age
sns.histplot(df['Age'], kde=True)
plt.show()

# 2. Countplot of Purchased
sns.countplot(x='Purchased', data=df)
plt.show()
```

# Bivariate

```python
# 1. Scatter Age vs. Salary
sns.scatterplot(x='Age', y='Salary', data=df)
plt.show()

# 2. Boxplot Salary by Purchased
sns.boxplot(x='Purchased', y='Salary', data=df)
plt.show()
```
